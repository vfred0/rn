# Visualización del modelo y la superficie del error

El siguiente código carga un conjunto de datos, y luego crea un modelo de regresión lineal para predecir los valores de salida $y$ en base a los de entrada $x1$ y $x2$ y en base a la función $y= m_1.x_1 + m_2.x_2 + b$.

Probá modificando los parámetros $m_1$ y $m_2$ y $b$ del modelo, tratando de minimizar *a mano* el error. Observá como cambia la recta $m_1.x_1 + m_2.x_2 + b$ y qué tan bien se ajusta a los datos.  

En este caso, ya no podemos visualizar la función de error ya que hay 3 parámetros y por ende necesitaríamos 4 dimensiones (3 para los parámetros y una para el valor del error en base a ellos).



## Carga de datos

```python
data = rnutil.load_dataset_numpy("study_regression_2d_small.csv")

x1, x2, y = data[:, 0], data[:, 1], data[: ,2]
```

## Parámetros del modelo (probar valores entre -5 y 5)

```python
m1 = 0.44
m2 = -1.0205
b  = 2
```

## Calculo del error del modelo

```python
y_predicted = x1 * m1 + x2 * m2 + b
mean_error  = np.mean((y - y_predicted) ** 2)
```

```python
rnutil.plot_regresion_lineal(
    [m1,m2],
    b,
    x1,
    x2,
    y,
    "x1 (Horas estudiadas)",
    "x2 (Promedio)",
    "y (Nota)",
    title="Modelo Final"
)
plt.show()
```
